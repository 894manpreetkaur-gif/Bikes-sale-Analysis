## 2. `Date`: Keep only date

```python
df["Date"] = pd.to_datetime(df["Date"]).dt.date
```

```python
df["Date"].unique()
```

```python
df["Date"].dtype
```

```python
df["Date"].head()
```

## 3. `Day`:- Need Imputing

```python
df.describe()
```

```python
df.fillna({"Day" : df["Day"].median()}, inplace=True)
```

```python
df.isnull().sum()
```

```python
df["Day"].unique()
```

## 4. `Month`:- Spelling variation

```python
df["Month"] =df["Month"].replace({
    "decmber":"December"
})
```

```python
df["Month"].unique()
```

## 7.`Age Group`:- Need imputing


```python
df.describe(include="all")
```

```python
df.fillna({"Age_Group" : df["Age_Group"].mode()[0]}, inplace = True)
```

## 9. `Country`:- Extra Space 

```python
df["Country"] = df["Country"].replace({
    " United States":"United States", "United  States": "United States", "United States ": "United States"
})
```

## 13. `Product_Description`:- Need imputing

```python
a = df["Product_Description"].mode()[0]
a
```

```python
df.fillna({"Product_Description" : df["Product_Description"].mode()[0]}, inplace = True)
```

## 14. ` Order_Quantity`:-  Need imputing

```python
df.fillna({"Order_Quantity" : df["Order_Quantity"].median()}, inplace = True)
```

```python
df.isnull().sum()
```

 ## 15.`Unit_cost, Unit_Price, Profit , Coat`:- Extra space in column name

```python
df.rename(columns={
    " Unit_cost":"Unit_Cost",
    " Unit_Price ":"Unit_Price",
    " Profit ":"Profit", 
    " Cost ":"Cost"
}, inplace = True)
```

# Graphs:- Hidden Insights

### `fig1`:- Bikes purchase according to the age group

```python
#Pie Chart
fig1 = px.pie(df,values="Order_Quantity", names="Age_Group" , title = "Bikes Purchasing according the Age Group")
fig1.show()
```

##### This pie chart shows how bike purchases are distributed among different age groups.
##### Overall:- Most bikes are purchased by Adults and Young Adults, indicating these groups are the primary target customers for bike sales.

#### `fig2`:- Relationship between Customer_Gender and Profit

```python
#bar chart
plt.bar(df["Customer_Gender"], df["Profit"])
plt.xlabel("Customer_Gender")
plt.ylabel("Profit")
plt.show()
```

##### This chart shows the the total profit contribution from each gender group.
##### Overall:- Female customers generate significantly higher profit than male customers, making them the most valuable customer segment in this dataset.
